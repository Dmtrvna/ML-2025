#### Типы переменных

Номинальные переменные (не имеют порядка и сравнения):

*   Gender — пол (2 значения), без порядка;
*   City — город проживания, категории без порядка;
*   Profession — профессия студента, категории без порядка;
*   Dietary Habits — пищевые привычки (4 значения), скорее номинальная, но если значения — шкала от маленького к большему по уровню здоровья, то будет порядковой;
*   Have you ever had suicidal thoughts ? — были ли суицидальные мысли (2 значения), без порядка;
*   Family History of Mental Illness — семейная история психических заболеваний (2 значения);
*   Depression — наличие депрессии, 2 уникальных значения (0/1 or Yes/No) — целевая бинарная номинальная.

Порядковые переменные (с порядком, без смысла мат операций):

*   Academic Pressure — уровень учебного давления (6 значений), вероятно шкала от маленького к большему, упорядоченная;
*   Work Pressure — уровень рабочего давления (3 значения), тоже скорее шкала;
*   Study Satisfaction — удовлетворённость учёбой (6 значений), вероятно шкала;
*   Job Satisfaction — удовлетворённость работой (5 значений), вероятно шкала;
*   Sleep Duration — длительность сна (5 значений), скорее всего классификация, допустим "мало"/"нормально"/"много";
*   Degree — уровень обучения (28 значений), скорее номинальная, хотя может быть некоторый порядок (бакалавр, магистр, кандидат и т.п.), но без числовой операции — порядковая;
*   Financial Stress — уровень финансового стресса (6 значений), вероятно порядковая шкала.

Числовые непрерывные переменные (с порядком и смыслом мат операций):

*   Age — возраст, количественная непрерывная переменная;
*   CGPA — средний балл, числовая оценка — количественная непрерывная;
*   Work/Study Hours — часы работы и учёбы, количественная непрерывная.
