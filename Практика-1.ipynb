#### 1. **Гистограмма для числовых переменных**

**задача ⏰**

Описать вывод графиков.
- Является ли распределение нормальным или отлично от него?
- Концентрация большинства значений
- если есть категориальная переменная, то с помощью параметры hue попробовать предсказать значение категориальный переменной (пример с диоксидом серы и типом вина)

p.s. 3-5 самых показательных примеров.

```python
# Числовые переменные
numeric = ["Age", "CGPA", "Work/Study Hours"]

# Гистограмма для числовых переменных
fig, axes = plt.subplots(1, len(numeric), figsize=(6 * len(numeric), 4))

for ax, col in zip(axes, numeric):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(f'Распределение {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Частота')

plt.tight_layout()
plt.show()
```

График распределения Age (Возраст) показывает, что большинство людей молодые, в основном от 18 до 35 лет. График выглядит “рвано”, с сильными перепадами, что не похоже на нормальное распределение. Таким образом, в дальнейшем мы будем исследовать наличие депресии у студентов в возрасте от 18 до 35 лет.

График CGPA (Средний балл) показывает, что у большинства опрошеннных средний балл довольно высокий и составляет от 5 до 10, почти никто не получает очень низкие оценки. График не ровный, а с “лесенкой”, значит оценки часто повторяются и нормального распределения тут нет.

В графике Work/Study Hours (Часы работы/учёбы) мы видим много пробелов, значит нрмального распределения тут нет. По графику можно сделать вывод, что больше всего людей работает/учится по 8-12 часов.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Распределение Work/Study Hours по Sleep Duration
sns.histplot(data=df, x="Work/Study Hours", hue="Sleep Duration", kde=True, multiple="stack", ax=axes[0])
axes[0].set_title("Часы учёбы/работы по длительности сна")

# Распределение Work/Study Hours по Study Satisfaction
sns.histplot(data=df, x="Work/Study Hours", hue="Study Satisfaction", kde=True, multiple="stack", ax=axes[1])
axes[1].set_title("Часы учёбы/работы по удовлетворённости")

# Распределение CGPA по Study Satisfaction
sns.histplot(data=df, x="CGPA", hue="Study Satisfaction", kde=True, multiple="stack", ax=axes[2])
axes[2].set_title("CGPA по удовлетворённости обучением")

plt.tight_layout()
plt.show()
```

Наиболее часто наблюдается снижение длительности сна при увеличении часов работы/учёбы. Так же при увеличении часов работы/учёбы наблюдается снижение уровня удовлетворённости обучением. Наиболее часто высокий средний балл снижает уровень удовлетворённости обучением.

#### 4. **Постройте корреляционную матрицу для всех числовых переменных и визуализируйте её с помощью тепловой карты**
